```python
%matplotlib inline

from nuscenes import NuScenes

nusc = NuScenes(version='v1.0-mini', dataroot='/home/haomo/Documents/nuScenes-mini', verbose=True)
```

```python
nusc.list_scenes()
```

```python
my_scene = nusc.scene[0]
my_scene
```

```python
first_sample_token = my_scene['first_sample_token']
first_sample_token
```

```python
my_sample = nusc.get('sample', first_sample_token)
my_sample
```

```python
sensor = 'CAM_FRONT'
cam_front_data = nusc.get('sample_data', my_sample['data'][sensor])
cam_front_data
```

```python
nusc.render_sample_data(cam_front_data['token'])
# 显示原图
```

```python
my_annotation_token = my_sample['anns'][22]
my_annotation = nusc.get('sample_annotation', my_annotation_token)
my_annotation
nusc.render_annotation(my_annotation_token)
```

```python
my_instance = nusc.instance[322]
my_instance
```

```python
instance_token = my_instance['token']
instance_token
nusc.render_instance(instance_token)
```

```python
print("First annotated sample of this instance:")
nusc.render_annotation(my_instance['first_annotation_token'])
```

```python
print("Last annotated sample of this instance:")
nusc.render_annotation(my_instance['last_annotation_token'])
```
